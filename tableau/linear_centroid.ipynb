```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets

#iris = datasets.load_iris()
#X=iris.data
#X

df = pd.read_csv('centroid_linestring.csv', header=0, sep=',')

X = df.iloc[:,[2,3]].values

X
```

```python
from sklearn.cluster import AgglomerativeClustering

ml=AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='ward')

ml.fit_predict(X)

plt.scatter(X[:,0],X[:,1],c=ml.fit_predict(X))
```

```python
import scipy.cluster.hierarchy as sch


dis=sch.linkage(X,metric='euclidean',method='ward')

plt.figure(figsize=(10, 7))
sch.dendrogram(dis)
plt.title('Hierarchical Clustering')
plt.show()
```

```python
from scipy.cluster.hierarchy import ward, fcluster
df2 = pd.DataFrame(fcluster(dis, t=0.25, criterion='distance'))
df2
df2[[0]].groupby([0]).count()
```

```python
df2.to_csv('centroid_hierarchy_output.csv')
```
